### Data Cleaning
If you look at the data we can see some issues,

```python
# remove "*", NaN, add missing sales data to "missing" dataframe 
# sales = sales['Country'].fillna('Unknown')    ## Do not use this method as it overwrites the dataframe


sales['Item Type'] = sales['Item Type'].apply(lambda x:str(x).replace('*',''))
sales = sales.fillna({'Country': 'unknown',  'Region' : 'unknown'})

missing = sales[sales.isna().any(axis=1)]
# missing = pd.DataFrame(x) ##creating new dataframe with null values
# missing

# sales = sales.dropna(axis = 0, how ='any')  ## Removing null values from existing dataframe
# sales = sales.reset_index(drop=True)

# raise NotImplementedError()
```
